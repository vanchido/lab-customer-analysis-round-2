```python
#For the numerical variables, check the multicollinearity between the features. 
#Please note that we will use the column total_claim_amount later as the target variable.

X = numerical.drop(['Total Claim Amount'], axis=1)
y = numerical['Total Claim Amount']

# with 'add_constant' we force the model to have an intercept
X = sm.add_constant(X)

model = sm.OLS(y,X).fit() # to train the model

print(model.summary())
```

```python
X = numerical.drop(['Total Claim Amount'], axis=1) 
y = numerical['Total Claim Amount'] 

lm = LinearRegression() 
lm.fit(X,y) 
```

```python
predictions = lm.predict(X) 

rmse = mean_squared_error(y, predictions, squared=False) 
mae = mean_absolute_error(y, predictions)
# r2 gives you an overall estimate of how good the model is, between 0 (lowest) and 1 (highest)
# rmse is good to see if the errors are big, the values returned are in the same unit as the target
print("R2_score:", round(lm.score(X,y),2)) 
print("RMSE:", rmse)
print("MAE:", mae)
```

```python
np.percentile(nums['Total Claim Amount'],25)
```

```python
upper_limit = np.percentile(nums['Total Claim Amount'],75) + 1.5*iqr
lower_limit = np.percentile(nums['Total Claim Amount'],25) - 1.5*iqr
upper_limit, lower_limit
```

```python
data = nums[(nums['Total Claim Amount']>lower_limit) & (nums['Total Claim Amount']<upper_limit)]
data
```

```python
sns.pairplot(data)
```

```python
for column in data.columns:
    sns.distplot(data[column])
    plt.show()
```

```python
#log transformation
def log_lmbda(x):
    return np.log(x) if np.isfinite(np.log(x)) else np.NAN

def log_transform_clean(x):
    return x if np.isfinite(x) else np.NAN
```

```python
data1 = data.copy()
data1.columns
```

```python
data1 = data.copy()

for column in data1.columns:
    data1[column] = data1[column].apply(log_lmbda)
    data1[column] = data1[column].fillna(np.mean(data1[column]))
    sns.displot(data[column])
    plt.show()
```

```python
data1 = data.copy()

for column in data1.columns:
    data1[column] = data1[column].apply(log_transform_clean)
    data1[column] = data1[column].fillna(np.mean(data1[column]))
    sns.displot(data[column])
    plt.show()
```

```python
data1
```

```python
#BoxCox transformation
data1 = data.copy()
for column in data1.columns:
    data1[column] = np.where(data[column]<=0, np.NAN, data[column])
    data1[column] = data1[column].fillna(data[column].mean())
    transformed_data = stats.boxcox(data1[column])
    sns.displot(transformed_data)
    plt.show()
```
